# Lab 6: Text-to-Speech (TTS)

- Level 1: Rule-based
- Level 2: Deep Learning
- Level 3: Few-shot Voice Cloning

```python
import pyttsx3
import numpy as np
from scipy.io import wavfile

engine = pyttsx3.init()
engine.setProperty('rate', 150)
text = 'Xin chao, day la Text-to-Speech level 1.'
engine.say(text)
engine.runAndWait()
print(' Level 1 Complete')
```

## Level 2: Deep Learning with gTTS

```python
from gtts import gTTS
text = 'Hello, this is Level 2 using Google Deep Learning TTS.'
tts = gTTS(text=text, lang='en')
tts.save('level2_output.mp3')
print(' Level 2 Complete')
```
